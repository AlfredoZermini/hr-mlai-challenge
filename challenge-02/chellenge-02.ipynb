```python
import pandas as pd
```

```python
# Load CSV file using pandas
csv_file_path = 'dataset.csv'  # Replace with the path to your CSV file

# Read header separately with a different delimiter
#df = pd.read_csv(csv_file_path, sep='[;]+')
df = pd.read_csv(csv_file_path, sep=';')

# display all columns
pd.set_option('display.max_columns', None)

# remove NaNs
df_cleaned = df.dropna(axis=1, how='all')
display(df_cleaned)
```

```python
# transform into numbers
df_numeric = df_cleaned.apply(pd.to_numeric, errors='coerce')

# delete -99, they do no look meaningful as they are temperatures and 0s, as they refer to humidity values
below_threshold = df_numeric.columns[(df_numeric <= 0).any()]
df_cleaned2 = df_cleaned.drop(columns=below_threshold)
display(df_cleaned2)
```
